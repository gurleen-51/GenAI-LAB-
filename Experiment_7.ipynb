# Experiment 7: Model Evaluation (Validation)

### Aim
Evaluate accuracy and loss metrics on a held-out validation set to monitor for overfitting and generalization.

### Concept
Model Validation ensures that the learned features are representative of the global distribution, not just the training noise. We use Train-Val-Test splits to protect against data leakage.

### Mathematical Foundations
$$\text{Generalization Gap} = Error_{val} - Error_{train}$$

### System Architecture
*   Data Partitioning: Split into 70% Train, 15% Val, 15% Test.
*   Metric Tracking: Monitor Loss/Accuracy across epochs.
*   Comparison: Bar chart of final validation metrics.
*   Analysis: Identify overfitting point where Val Loss increases.

---

```python
import matplotlib.pyplot as plt
import numpy as np

epochs = np.arange(1, 21)
train_loss = np.exp(-epochs/5) + 0.1
val_loss = train_loss + 0.05 * (epochs/10)**2

plt.figure(figsize=(8, 5))
plt.plot(epochs, train_loss, label='Train Loss')
plt.plot(epochs, val_loss, label='Val Loss', linestyle='--')
plt.axvline(12, color='r', linestyle=':', label='Overfit Point')
plt.title("Overfitting Analysis")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
