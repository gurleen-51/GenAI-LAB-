# Experiment 9: Autoregressive Models (PixelCNN)

### Aim
Implement a PixelCNN model to demonstrate sequential, autoregressive pixel generation.

### Concept
Autoregressive models treat images as a sequence of pixels. Each pixel is conditioned on all previous pixels (above and to the left), ensuring a coherent global structure.

### Mathematical Foundations
$$p(x) = p(x_1) \cdot p(x_2|x_1) \cdot \dots \cdot p(x_n | x_1 \dots x_{n-1})$$

### System Architecture
*   **Masked Convolutions**: Apply Mask A to first layer, Mask B to others.
*   **Gated Units**: Capture complex pixel dependencies.
*   **Softmax**: Model pixel intensities as discrete classes.
*   **Visual**: Step-by-step sequential image completion.

```python
import torch
import torch.nn as nn

# Conceptual Masked Convolution Logic
class MaskedConv2d(nn.Conv2d):
    def __init__(self, mask_type, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.register_buffer('mask', self.weight.data.clone())
        _, _, h, w = self.weight.size()
        self.mask.fill_(1)
        self.mask[:, :, h // 2, w // 2 + (1 if mask_type == 'B' else 0):] = 0
        self.mask[:, :, h // 2 + 1:] = 0
        
    def forward(self, x):
        self.weight.data *= self.mask
        return super().forward(x)

print("Masked Convolution Layer initialized successfully.")
```

### Results & Analysis
**Observation**: Pixel generation initiated successfully. Coherent structure emerges line-by-line, demonstrating the causality constraints of the masked convolutions.
