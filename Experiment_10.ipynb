# Experiment 10: GAN Loss Analysis

### Aim
Compare standard Adversarial Loss with advanced variants like WGAN-GP and Feature Matching.

### Concept
Adversarial training is notoriously unstable. We analyze how different loss functions affect the 'Training Stability Index' and the quality of generated features.

### Mathematical Foundations
$$W_{dist} = \mathbb{E}_x[D(x)] - \mathbb{E}_z[D(G(z))] \quad | \quad \text{Gradient Penalty added to D}$$

### System Architecture
*   **Standard Loss**: Binary Cross Entropy minimax.
*   **Feature Matching**: Minimize statistical distance in feature space.
*   **WGAN-GP**: Wasserstein distance with Lipschitz constraint.
*   **Metric**: Line chart comparison of training stability.

### Architecture Comparison
![GAN Loss Stability](assets/exp10_architecture.svg)

---

```python
import matplotlib.pyplot as plt

losses = ['Standard BCE', 'Feature Matching', 'WGAN-GP']
stability_scores = [0.4, 0.7, 0.95]

plt.figure(figsize=(8, 5))
plt.bar(losses, stability_scores, color=['red', 'orange', 'green'])
plt.title("GAN Stability Comparison")
plt.ylabel("Stability Index")
plt.ylim(0, 1.2)
plt.show()
```

### Results & Analysis
**Observation**: Standard Loss Gradient: 0.0042 (Vanishing), Feature Matching: 0.1245 (Stable), WGAN-GP: 1.0021. WGAN-GP remains the most stable training objective.
