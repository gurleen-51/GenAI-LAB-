# Experiment 1: Probability Distributions

### Aim
Generate random samples from various probability distributions (Normal, Uniform) to analyze density patterns and statistical foundations.

### Concept
In Deep Learning, weight initialization and data augmentation rely on understanding probability density. We study the Normal and Uniform distributions as they are the most fundamental building blocks for stochastic processes.

### Mathematical Foundations
- **Normal PDF**: $f(x|\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$
- **Uniform PDF**: $f(x|a,b) = \frac{1}{b-a}$ for $a \le x \le b$

### System Architecture
*   Define distribution parameters ($\mu=0, \sigma=1$ for Gaussian; range $[-2, 2]$ for Uniform).
*   Generate 5,000 synthetic samples using NumPy's random module.
*   Apply Seaborn darkgrid styling for professional visualization.
*   Visualize using Histograms with Kernel Density Estimation (KDE).

---

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')

# 1. Normal Distribution (μ=0, σ=1)
normal_samples = np.random.normal(0, 1, 5000)

# 2. Uniform Distribution (-2, 2)
uniform_samples = np.random.uniform(-2, 2, 5000)

print(f"Normal Mean: {np.mean(normal_samples):.3f}, Std: {np.std(normal_samples):.3f}")
print(f"Uniform Min: {np.min(uniform_samples):.3f}, Max: {np.max(uniform_samples):.3f}")

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(normal_samples, kde=True, color='purple')
plt.title('Normal Distribution')

plt.subplot(1, 2, 2)
sns.histplot(uniform_samples, kde=True, color='teal')
plt.title('Uniform Distribution')
plt.show()
```
