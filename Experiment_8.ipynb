# Experiment 8: Variational Autoencoders (VAE)

### Aim
Build an Encoder-Decoder VAE and visualize the continuous latent space manifold.

### Concept
VAEs use variational inference to map data to a latent distribution. By sampling from the learned manifold, we can generate new, novel samples that share characteristics with the training set.

### Mathematical Foundations
$$L(\theta, \phi; x) = -D_{KL}(q_\phi(z|x) || p_\theta(z)) + \mathbb{E}_q[\log p_\theta(x|z)]$$

### System Architecture
*   **Encoder**: Input(784) $\rightarrow$ Dense(400) $\rightarrow$ [Mu(20), LogVar(20)].
*   **Reparameterization**: Stochastic sampling in 20D latent space.
*   **Decoder**: Latent(20) $\rightarrow$ Dense(400) $\rightarrow$ Output(784).
*   **Visual**: 15x15 grid sampling of the digit manifold.

```python
import torch
import torch.nn as nn

# Conceptual VAE Reparameterization Demo
def reparameterize(mu, logvar):
    std = torch.exp(0.5 * logvar)
    eps = torch.randn_like(std)
    return mu + eps * std

mu = torch.tensor([0.0, 0.0])
logvar = torch.tensor([0.1, 0.1])
z = reparameterize(mu, logvar)
print(f"Sampled Latent Vector z: {z}")
```
